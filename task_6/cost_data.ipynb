```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Завантаження попередньо підготовленого DataFrame з SCOR

# Попередня обробка
df = df_selected[
    (df_selected["Вид сільськогосподарської продукції"] != "Не застосовується") &
    (df_selected["Територіальний розріз"] != "Україна")
].dropna(subset=["2018", "2019", "2020"])

# Нормалізація
for year in ["2018", "2019", "2020"]:
    min_val = df[year].min()
    max_val = df[year].max()
    df[year + "_norm"] = (df[year] - min_val) / (max_val - min_val)

# Обчислення SCOR
df["SCOR"] = df[["2018_norm", "2019_norm", "2020_norm"]].mean(axis=1)

# Побудова графіків

# 1. Теплова карта по SCOR
pivot_df = df.pivot_table(index="Вид сільськогосподарської продукції",
                          columns="Територіальний розріз",
                          values="SCOR")

plt.figure(figsize=(14, 10))
sns.heatmap(pivot_df, cmap="YlGnBu", linewidths=0.5)
plt.title("SCOR (нормалізовані витрати) по культурах і регіонах")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
