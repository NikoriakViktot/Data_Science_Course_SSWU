```python

# Попередня обробка
df = dff[
    (dff["Вид сільськогосподарської продукції"] != "Не застосовується") &
    (dff["Територіальний розріз"] != "Україна")
].dropna(subset=["2018", "2019", "2020"])

# Нормалізація
for year in ["2018", "2019", "2020"]:
    min_val = df[year].min()
    max_val = df[year].max()
    df[year + "_norm"] = (df[year] - min_val) / (max_val - min_val)

# Обчислення SCOR
df["SCOR"] = df[["2018_norm", "2019_norm", "2020_norm"]].mean(axis=1)

# Побудова графіків

# 1. Теплова карта по SCOR
pivot_df = df.pivot_table(index="Вид сільськогосподарської продукції",
                          columns="Територіальний розріз",
                          values="SCOR")

plt.figure(figsize=(14, 10))
sns.heatmap(pivot_df, cmap="YlGnBu", linewidths=0.5)
plt.title("SCOR (нормалізовані витрати) по культурах і регіонах")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

# Вибираємо тільки потрібні колонки
columns = ["Вид сільськогосподарської продукції", "Територіальний розріз", "2018", "2019", "2020"]
df_crop = df[columns].copy()

# Створюємо pivot-таблицю: культури × регіони → значення = 2020 (наприклад)
pivot = df_crop.pivot_table(index="Вид сільськогосподарської продукції",
                            columns="Територіальний розріз",
                            values="2019",  # або '2018' чи '2019'
                            aggfunc='mean')

# Карта пропущених значень
plt.figure(figsize=(16, 8))
sns.heatmap(pivot.isnull(), cmap="YlGnBu", cbar=False, linewidths=0.5, linecolor='gray')
plt.title("Пропущені значення по культурах і регіонах ")
plt.xlabel("Регіон")
plt.ylabel("Культура")
plt.tight_layout()
plt.show()
```

```python
# Карта заміни назв культур
target_crops_map = {
    "Картопля": "Картопля",
    "Кукурудза на зерно": "Кукурудза",
    "Пшениця": "Пшениця",
    "Насіння соняшнику": "Соняшник",
    "Боби сої": "Соя",
    "Ячмінь": "Ячмінь"
}

# Карта для заміни назв регіонів
region_translation = {
    "Івано-Франківська": "Ivano-Frankivska oblast",
    "Вінницька": "Vinnytska oblast",
    "Волинська": "Volynska oblast",
    "Дніпропетровська": "Dnipropetrovska oblast",
    "Донецька": "Donetska oblast",
    "Житомирська": "Zhytomyrska oblast",
    "Закарпатська": "Zakarpatska oblast",
    "Запорізька": "Zaporizka oblast",
    "Кіровоградська": "Kirovohradska oblast",
    "Київська": "Kyivska oblast",
    "Луганська": "Luhanska oblast",
    "Львівська": "Lvivska oblast",
    "Миколаївська": "Mykolaivska oblast",
    "Одеська": "Odeska oblast",
    "Полтавська": "Poltavska oblast",
    "Рівненська": "Rivnenska oblast",
    "Сумська": "Sumska oblast",
    "Тернопільська": "Ternopilska oblast",
    "Харківська": "Kharkivska oblast",
    "Херсонська": "Khersonska oblast",
    "Хмельницька": "Khmelnytska oblast",
    "Черкаська": "Cherkaska oblast",
    "Чернівецька": "Chernivetska oblast",
    "Чернігівська": "Chernihivska oblast"
}

# Фільтрація по культурі і регіону
target_crops = list(target_crops_map.keys())
df_filtered = dff[
    (dff["Вид сільськогосподарської продукції"].isin(target_crops)) &
    (dff["Територіальний розріз"].isin(region_translation.keys()))
].copy()

# Заміна назв культур на уніфіковані
df_filtered["Культура"] = df_filtered["Вид сільськогосподарської продукції"].map(target_crops_map)

# Заміна назв регіонів
df_filtered["Region"] = df_filtered["Територіальний розріз"].map(region_translation)

# Нормалізація по кожному року
for year in ["2018", "2019", "2020"]:
    min_val = df_filtered[year].min()
    max_val = df_filtered[year].max()
    df_filtered[year + "_norm"] = (df_filtered[year] - min_val) / (max_val - min_val)

# Обчислення SCOR
df_filtered["SCOR"] = df_filtered[["2018_norm", "2019_norm", "2020_norm"]].mean(axis=1)

# Створення таблиці для теплової карти
pivot_df = df_filtered.pivot_table(index="Культура", columns="Region", values="SCOR")

print(pivot_df)
```
