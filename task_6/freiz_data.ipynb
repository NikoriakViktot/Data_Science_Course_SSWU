```python
import matplotlib.pyplot as plt

def plot_crop_dynamics(df, crop_name):
    crop_df = df[df["Вид сільськогосподарських культур"] == crop_name]
    crop_df = crop_df[crop_df["Територіальний розріз"] != "Україна"]

    if crop_df.empty:
        print(f"Культура '{crop_name}' не знайдена у даних.")
        return

    pivot_df = crop_df.groupby("Територіальний розріз")[["2019", "2020", "2021", "2022", "2023"]].sum().T

    plt.style.use("dark_background")
    pivot_df.plot(figsize=(14, 6), marker='o')
    plt.title(f"Динаміка внесення добрив для {crop_name.lower()} (по регіонах, без України)")
    plt.ylabel("Обсяг (кг у діючій речовині)")
    plt.xlabel("Рік")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Оновлений список культур
target_crops = [
    "Пшениця",
    "Кукурудза",
    "Соняшник",
    "Соя",
    "Культури зернові та зернобобові",
    "Коренеплоди та бульби їстівні з високим вмістом крохмалю та інуліну"
]

# Побудова графіків
for crop in target_crops:
    plot_crop_dynamics(df, crop)

```

```python

# Фільтрація лише на "Усього" у діючій речовині
df_filtered = df[
    (df["Підвид засобу"] == "Усього") &
    (df["Тип маси"] == "У діючій речовині") &
    (df["Вид речовини"] == "Усього")
].copy()

# Цільові культури
target_crops = [
    "Пшениця", "Кукурудза", "Соняшник", "Соя",
    "Культури зернові та зернобобові",
    "Коренеплоди та бульби їстівні з високим вмістом крохмалю та інуліну"
]

# Функція заповнення пропусків середнім по регіону
def fill_missing_with_mean(df, crop_name):
    crop_df = df[
        (df["Вид сільськогосподарських культур"] == crop_name) &
        (df["Територіальний розріз"] != "Україна")
    ].copy()

    cols = ["Територіальний розріз", "2019", "2020", "2021", "2022", "2023"]
    crop_df = crop_df[cols]

    # Групування і заповнення NaN середнім по регіону
    crop_df_grouped = crop_df.groupby("Територіальний розріз").transform(lambda x: x.fillna(x.mean()))
    crop_df_grouped["Культура"] = crop_name
    crop_df_grouped["Територіальний розріз"] = crop_df["Територіальний розріз"].values

    return crop_df_grouped

# Створення заповненої таблиці
filled_tables = []
for crop in target_crops:
    try:
        filled_tables.append(fill_missing_with_mean(df_filtered, crop))
    except Exception as e:
        print(f" Помилка для '{crop}': {e}")

df_filled = pd.concat(filled_tables, ignore_index=True)

def compute_log_scor_from_filled(df):
    scor_tables = []
    cols_years = ['2019', '2020', '2021', '2022', '2023']

    for crop in df["Культура"].unique():
        sub_df = df[df["Культура"] == crop].copy()

        # Логарифмування з додаванням 1 — через apply
        log_df = sub_df[cols_years].apply(lambda col: np.log(col + 1))

        # Вивід діапазону
        print(f" ЛОГ-Мінімум для {crop}:\n", log_df.min())
        print(f" ЛОГ-Максимум для {crop}:\n", log_df.max())

        # Мін–макс нормалізація
        log_norm = (log_df - log_df.min()) / (log_df.max() - log_df.min())

        # Додати колонки
        log_norm["Культура"] = crop
        log_norm["Територіальний розріз"] = sub_df["Територіальний розріз"].values

        scor_tables.append(log_norm)

    return pd.concat(scor_tables, ignore_index=True)

#Викликаємо лог-нормалізовану SCOR
scor_df = compute_log_scor_from_filled(df_filled)

#  Округлення і збереження
scor_df[['2019', '2020', '2021', '2022', '2023']] = scor_df[['2019', '2020', '2021', '2022', '2023']].round(4)
scor_df = scor_df[["Культура", "Територіальний розріз", "2019", "2020", "2021", "2022", "2023"]]
scor_df.to_csv("scor_table_log_norm.csv", index=False)

print(scor_df.head(100))
```
